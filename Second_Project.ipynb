```python
import pandas as pd
import numpy as np

# 1. טעינת הנתונים (הקובץ הזה דורש קידוד לטיני בד"כ)
try:
    df = pd.read_csv('DataCoSupplyChainDataset.csv', encoding='ISO-8859-1')
    print("✅ הקובץ נטען בהצלחה!")
except FileNotFoundError:
    print("❌ שגיאה: הקובץ לא נמצא בתיקייה. וודא שהשם זהה.")

# 2. ניקוי שמות העמודות (הפיכתם לקלים לעבודה: בלי רווחים, הכל באותיות קטנות)
df.columns = [col.lower().replace(' ', '_').replace('(', '').replace(')', '') for col in df.columns]

# 3. המרת עמודות תאריך לפורמט של פייתון (קריטי לניתוח זמנים)
df['order_date_dateorders'] = pd.to_datetime(df['order_date_dateorders'])
df['shipping_date_dateorders'] = pd.to_datetime(df['shipping_date_dateorders'])

# 4. יצירת מדד חדש ומנצח: "הפרש ימי משלוח" (כמה ימים איחרנו בפועל לעומת התכנון)
df['delivery_delay'] = df['days_for_shipping_real'] - df['days_for_shipment_scheduled']

# הצגת המידע
print(f"בדאטה יש {df.shape[0]} שורות ו-{df.shape[1]} עמודות.")
print("\n--- מדדי איחורים ראשוניים (בממוצע) ---")
print(df[['delivery_delay', 'order_profit_per_order']].mean())
```
